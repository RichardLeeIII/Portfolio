```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

data = []

for page in range(2,5):
    URL = "https://suumo.jp/jj/bukken/ichiran/JJ012FC001/?ar=030&bs=011&cn=9999999&cnb=0&ekTjCd=&ekTjNm=&et=9999999&fw2=&kb=1&kt=9999999&mb=0&md=2&mt=9999999&sc=13101&sc=13102&sc=13103&sc=13104&sc=13105&sc=13113&sc=13106&sc=13107&sc=13118&sc=13121&sc=13122&sc=13123&sc=13109&sc=13110&sc=13111&sc=13112&sc=13114&sc=13115&sc=13120&sc=13116&sc=13117&sc=13119&scTemp=13101&scTemp=13102&scTemp=13103&scTemp=13104&scTemp=13105&scTemp=13113&scTemp=13106&scTemp=13107&scTemp=13118&scTemp=13121&scTemp=13122&scTemp=13123&scTemp=13109&scTemp=13110&scTemp=13111&scTemp=13112&scTemp=13114&scTemp=13115&scTemp=13120&scTemp=13116&scTemp=13117&scTemp=13119&ta=13&tj=0&po=0&pj=1&pc=100&page=" + str(page)
    response = requests.get(URL)
    soup = BeautifulSoup(response.content, 'lxml')
    print(f"Processing page {page}, Status Code: {response.status_code}")

    # Extract the data from the soup object
    results = soup.find_all('dd')
    add = [result.text for result in results]
    data.extend(add[8:736])

# Create a pandas DataFrame from the extracted data
df = pd.DataFrame(data)

# Print the DataFrame
print(df)




```
