```python
from bs4 import BeautifulSoup 
import requests
import urllib3
```

```python
URL = "https://www.worldometers.info/world-population/population-by-country/"
response = requests.get(URL) 
soup = BeautifulSoup(response.content,'lxml')
print("confirm",response.status_code)
```

```python
print(soup)
```

```python
with open("weatherinfo2.html","w",encoding="utf8") as f:
    f.write(str(soup))
```

```python
with open("weatherinfo2.html","r",encoding="utf8") as f:
    doc = BeautifulSoup(f, "html.parser")
    
print(doc)
```

```python
print(doc.prettify())
```

```python
tag = doc.title
print(tag.string)
```

```python
table=doc.find('table')
table.prettify
```

```python
headers=[]

for i in table.find_all('th'):
    title=i.text
    headers.append(title)
```

```python
headers
```

```python
import pandas as pd
df = pd.DataFrame(columns=headers)

df
```
